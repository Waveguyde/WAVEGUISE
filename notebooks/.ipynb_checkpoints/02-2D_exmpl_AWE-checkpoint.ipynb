# WAVEGUISE applied to AWE data
WAVEGUISE is applied to an AWE swath over Crozet island in the south Indian ocean. Orbit 3175

```python
import sys
sys.path.append("/home/r/Robert.Reichert/juwavelet")
import juwavelet.transform as transform
from juwavelet import utils
sys.path.append('../scripts')
import utils_2d as two

import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import copy
from scipy import stats
import matplotlib.colors as mcolors
from matplotlib import cm

plt.style.use("../latex_default.mplstyle")
```

```python
data_path = '../data/awe_l3c_q20_2024165T2120_03175_v01.nc'
ds = xr.open_dataset(data_path)
```

## Data Preprocessing
Before the observational data set is ready for the wavelet transform, we must take care of the data boundaries. These non-periodic boundaries can create strong frequency responses and later on fuck up the wavefield segmentation. The first step is to use a least square polynomial 2D fit of max_order and subtract the result. In a second step, the zeroth and first Fourier components are also subtracted. In a final step, a Tukey window is applied to smoothly enforce the signal to zero at the edges. 

```python
orig =ds["Radiance"].isel(time=0,x_along_track=slice(3000,3500),y_cross_track=slice(60,350))
BI_high, BI_low = two.BG_removal(orig.values,max_order=1)
dx, dy = 2000, 2000
nx, ny = orig['x_along_track'].size, orig['y_cross_track'].size
y = np.linspace(0,(ny-1)*dy,ny)*1e-3
x = np.linspace(0,(nx-1)*dx,nx)*1e-3
print(nx, ny)

nmask  = orig.shape
ntap   = int(0.1*np.min(nmask))
signal = utils.smooth_edges(BI_high, ntap, window='tukey')
```

## Wavelet Transform
When we do the juwavelet transform we need to choose the number of scales $js$ and orientations $jt$. The smallest scales is defined as $s_0=2*\delta x$ as suggested by Torrence and Compo (1998). The number of scales is taken to depend on the data dimensions and since the first Fourier component is already subtracted, we focus here on wavelengths shorter than that. For more details on juwavelet, please have a look at Ungermann and Reichert (2025).

```python
s0 = 2 * dx
dj = 1 / 8
js = int(1/dj*np.log2(min([nx,ny])/2*dx/s0))
jt = 21
print(js, jt)

result = transform.decompose2d(signal,dx,dy,s0,dj,js,jt,aspect=1,
                               nxpad=None,nypad=None,opts={'param': 2*np.pi},
                               mode="scaled",dtype=np.complex128)

print(result["period"][0],result["period"][-1])
```

## Red Noise Filtering
Before we continue with the segmentation, it is very useful to strip off any red noise. This is done by normalizing the wavelet power spectrum using the median and scaled median absolute deviation (sMAD) and setting all wavelet coefficients to zero where the normalized WPS is below a certain threshold (8 seems to be a good number in this application).

```python
cwt_copy   = copy.deepcopy(result)
# Compute the wavelet power spectrum
WPS        = np.abs(cwt_copy['decomposition'])**2
# Set wavelet coefficients to zero where we expect to find white noise
white_noise_threshold = 0.1
mask_1     = WPS < white_noise_threshold**2
cwt_copy['decomposition'][mask_1] = 0
WPS        = np.abs(cwt_copy['decomposition'])**2
# Compute med(WPS) and sMAD(WPS) for each scale independently!
median_WPS = np.median(WPS,axis=(1,2,3))
sMAD_WPS   = 1.4826*stats.median_abs_deviation(WPS,axis=(1,2,3))
# Compute the normalized WPS
WPS_scaled = (WPS-median_WPS[:,np.newaxis,np.newaxis,np.newaxis])/sMAD_WPS[:,np.newaxis,np.newaxis,np.newaxis]
# Set all wavelet coefficients to zero where the normalized WPS is below 8
WPS_noise_threshold = 8
mask       = WPS_scaled < WPS_noise_threshold
cwt_copy['decomposition'][mask]=0
# Reconstruct the wavefield without noise.
wavy_stuff  = transform.reconstruct2d(cwt_copy)
noisy_stuff = signal - wavy_stuff 
```

## Segmentation
In the following step, we apply the watershed segmentation algorithm to the filtered WPS. A crucial parameter is the prominence, which defines the height of a peak in the WPS comapred to its neighborhood. A suggestion for the prominence parameter would be the noise level of the underlying data. Depending on the prominence parameter we will find more or less wave packets. If there are too many, a way to reduce their number and focus on important ones is by using the function update_segments(). Therein wave packets are sorted according to their variance contribution and only as many packets are returned until the threshold variance is reached.

```python
WPS        = np.abs(cwt_copy['decomposition'])
prominence = 0.25
_seg       = two.wavefield_segmentation_2d(WPS,prominence,connectivity_order=4)
print(np.max(_seg))

seg = two.update_segments(WPS,_seg,threshold=0.95,mode='sum')
print(np.max(seg))
```

## Reconstruction
Now that we grouped wavelet coefficients in the spectral space, we reconstruct each wave packet and investigate their individual frequency content.

```python
recon_seg, amp_seg, kx_seg, ky_seg = two.recon_segments_2d_v2(cwt_copy,seg)
lh_seg, th_seg = two.kxky_2_lhtheta(kx_seg,ky_seg)
```
